```python
images=[]
for i in range(37):
    images.append(imread(str(i+1)+'.jpg'))
```

```python
img=[]
l=[]
a=[]
for i in range(37):
    img.append(images[i][1000:1100,450:650])
    a.append(img[i][60:80,150:175])
    l.append(a[i].mean())
```

```python
import numpy as np

figsize(15,5)
l=np.array(l)
ang=np.arange(0,370,10)
angg=np.arange(0,370,0.1)

subplot(121)
plt.plot(ang,l,'ko')
plt.plot(angg,l.max()*np.cos(np.radians(angg))**2,'b-')

plt.subplots(subplot_kw={'projection': 'polar'})
plt.plot(np.radians(ang),l,'k-o')
plt.plot(np.radians(angg),l.max()*np.cos(np.radians(angg))**2,'-')
plt.show()
```

```python
imr=[]
lr=[]
for i in range(37):
    aim=a[i].copy()
    aim[:,:,[1,2]]=0
    lr.append(aim.mean())
    imr.append(aim)
img=[]
lg=[]
for i in range(37):
    aim=a[i].copy()
    aim[:,:,[0,2]]=0
    lg.append(aim.mean())
    img.append(aim)
imb=[]
lb=[]
for i in range(37):
    aim=a[i].copy()
    aim[:,:,[0,1]]=0
    lb.append(aim.mean())
    imb.append(aim)
```

```python
figsize(15,5)
lr=np.array(lr)
ang=np.arange(0,370,10)
angg=np.arange(0,370,0.1)

subplot(121)
plt.plot(ang,lr,'ro')
plt.plot(angg,lr.max()*np.cos(np.radians(angg))**2,'b-')

plt.subplots(subplot_kw={'projection': 'polar'})
plt.plot(np.radians(ang),lr,'r-o')
plt.plot(np.radians(angg),lr.max()*np.cos(np.radians(angg))**2,'-')
plt.show()
```

```python
figsize(15,5)
lg=np.array(lg)
ang=np.arange(0,370,10)
angg=np.arange(0,370,0.1)

subplot(121)
plt.plot(ang,lg,'go')
plt.plot(angg,lg.max()*np.cos(np.radians(angg))**2,'b-')

plt.subplots(subplot_kw={'projection': 'polar'})
plt.plot(np.radians(ang),lg,'g-o')
plt.plot(np.radians(angg),lg.max()*np.cos(np.radians(angg))**2,'-')
plt.show()
```

```python
figsize(15,5)
lb=np.array(lb)
ang=np.arange(0,370,10)
angg=np.arange(0,370,0.1)

subplot(121)
plt.plot(ang,lb,'bo')
plt.plot(angg,lb.max()*np.cos(np.radians(angg))**2,'-')

plt.subplots(subplot_kw={'projection': 'polar'})
plt.plot(np.radians(ang),lb,'b-o')
plt.plot(np.radians(angg),lb.max()*np.cos(np.radians(angg))**2,'-')
plt.show()
```

```python
plt.subplots(subplot_kw={'projection': 'polar'})
plt.plot(np.radians(ang),lr/lr.max(),'r-o')
plt.plot(np.radians(ang),lg/lg.max(),'g-o')
plt.plot(np.radians(ang),lb/lb.max(),'b-o')
plt.plot(np.radians(angg),np.cos(np.radians(angg))**2,'-')
```
