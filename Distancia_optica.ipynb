```python
im= imread("Referencia.tif")
print(im)
```

```python
imshow(im, cmap= "gray")
colorbar()
```

```python
sel1=255-im[0:458,75:100]
imshow(sel1)
colorbar()
```

```python
sel1bw=where(sel1>210,1,0)
imshow(sel1bw)
colorbar()
```

```python
from scipy.ndimage import label
sel1lb,n1=label(sel1bw)
imshow(sel1lb,cmap="jet")
colorbar()
print(n1)
```

```python
from scipy.ndimage import center_of_mass
ly=[]
for i in range(1,n1+1):
  ly.append(center_of_mass(sel1,sel1lb,i)[0])
print(ly)
```

```python
tp=[]
for i in range (1,n1-1):
  tp.append(3/(ly[i]-ly[i-1]))
print(tp)
```

```python
mean(tp)
```

```python
std(tp)
```
